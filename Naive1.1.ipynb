```python
import pandas as pd
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('/Users/lamprosganias/Desktop/Cybersecurity and Data Science/Β Εξάμηνο/Time-Series Analytics and Forecasting (CDS214)/Datasets/Daily-train.csv')
data.set_index('V1', inplace=True)
data
```

```python
data = data.iloc[0]
```

```python
data = data.dropna()  # drop null values
data = data.to_frame()
data
```

```python
data.plot(figsize=(20,10))
plt.show()
```

```python
fig = data.hist(figsize = (20,5))
```

```python
train = data
train
```

```python
train = train.reset_index(drop=True)
train
```

```python
train_len = len(train)
train_len
```

```python
test = pd.read_csv('/Users/lamprosganias/Desktop/Cybersecurity and Data Science/Β Εξάμηνο/Time-Series Analytics and Forecasting (CDS214)/Datasets/Daily-test.csv')
test.set_index('V1', inplace=True)
test
```

```python
test = test.iloc[0]
test
```

```python
test = test.dropna()  # drop null values
test = test.to_frame()
test
```

```python
test.set_index(pd.RangeIndex(start=train.index[-1]+1, stop=train.index[-1]+1+len(test)), inplace=True)
test
```

```python
y_hat_naive = test.copy()
y_hat_naive
```

```python
y_hat_naive['naive_forecast'] = train["D1"][train_len-1]
y_hat_naive['naive_forecast'].iloc[1:] = train['D1'][train_len-1]
y_hat_naive
```

```python
# plot the last 100 observations of train, test, and y_hat_naive
plt.figure(figsize=(20,5))
plt.grid()
plt.plot(train['D1'].tail(50), label='Train')
plt.plot(test['D1'].tail(50), label='Test')
plt.plot(y_hat_naive['naive_forecast'].tail(50), label='Naive forecast')
plt.legend(loc='best')
plt.title('Naive Method - Last 100 Observations')
plt.show()
```

```python
mse = mean_squared_error(test, y_hat_naive["naive_forecast"])
print(f'MSE: {mse:.2f}')

```

```python
import numpy as np 
rmse = np.sqrt(mean_squared_error(test, y_hat_naive['naive_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test - y_hat_naive['naive_forecast'])/test)*100,2)
results = pd.DataFrame({'Method':['Naive method'], 'MAPE': [mape], 'RMSE': [rmse]})
results = results[['Method', 'RMSE', 'MAPE']]
results
```
